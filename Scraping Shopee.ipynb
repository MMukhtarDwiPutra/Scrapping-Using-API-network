```python
arr = []
domain = "acc"
num = 358

for i in range(0,num,6):
    print(i)
#     urlreq = 'https://shopee.co.id/api/v2/item/get_ratings?filter=0&flag=1&itemid=2028506054&limit=6&&offset='+str(i)+'&shopid=18202344&type=4'
    urlreq = 'https://shopee.sg/api/v2/item/get_ratings?filter=1&flag=1&itemid=6231330923&limit=6&offset='+str(i)+'&shopid=182363386&type=1'
    
    # get response
    response = urllib.request.urlopen(urlreq)
    # load as json
    jresponse = json.load(response)
    # write to file as pretty print
    with open('data.json', 'w') as outfile:
        json.dump(jresponse, outfile, sort_keys=True, indent=4)

    import json
    from collections import OrderedDict

    path_to_file = "data.json";

    with open(path_to_file) as data_file:
        data = OrderedDict(json.load(data_file))

    # print(data)
    length = len(data["data"]["ratings"])


    for x in range(length):
#         print(data["data"]["ratings"][x]["comment"])
#         print()
        kalimat = " ".join(re.sub("(@[A-Za-z0-9]+)|([^0-9A-Za-z \t]) |(\w+:\/\/\S+)", " ",str(data["data"]["ratings"][x]["comment"])).split())
        string = nltk.word_tokenize(kalimat)
    
        if(len(string) > 3):
            if(data["data"]["ratings"][x]["rating_star"] > 3):
#                 print("skip")
                arr.append(kalimat+"; 1; "+domain+"; "+str(data["data"]["ratings"][x]["rating_star"]))
            else:
                arr.append(kalimat+"; 0; "+domain+"; "+str(data["data"]["ratings"][x]["rating_star"]))
      
    
file = open("hasil_scraping.txt", "w", encoding="utf-8")
for s in arr:
    file.writelines(s+"\n")
file.close()
```
